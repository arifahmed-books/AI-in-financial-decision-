### Import library

```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Customize

```python
# Customize parameters 
my_input = "0302 input_cost_data.csv"
my_correl = "0303 correlation.csv"
my_heatmap = "0303 correlation_heatmap.png"
```

### Load data and select columns 

```python
# Load the dataset
df = pd.read_csv(my_input)

# Select numeric columns only
numeric_columns = df.select_dtypes(include='number')
```

### Create, plot and save heatmap

```python
# Compute correlation matrix 
correlation_matrix = numeric_columns.corr()

# Plot heatmap 
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5,
    fmt=".2f"
)
plt.title("Correlation Heatmap of Cost Drivers and Overhead Cost")

# Save plot as image file 
plt.savefig(my_heatmap)
plt.show()
```

### Save correlation matrix and plot

```python
# Save the correlation matrix to a CSV file
correlation_matrix.to_csv(my_correl)

# Output path for reference 
print(f"Heatmap saved to: {my_heatmap}")
print(f"Correlation matrix saved to: {my_correl}")
```
